# Numpy

```python
# Biblioteca do Python
# Utilizadas para cálculos númericos em arrays (multidimencionais) e matrizes
# Permite a manipulação eficiente de dados complexos
# Data Science -> Tratamento dos Dados

import numpy as np

# Criar arrays de valores de valores aleatórios
array1 = np.random.randint(0,10,(3,3))
array2 = np.random.randint(0,10,(3,3))

print(array1)
print(array2)
```

```python
soma = array1 + array2
print("Soma: ",soma)

subtracao = array1 - array2
print("Subtração: ",subtracao)

multiplicacao = array1 * array2
print("Multiplicação: ",multiplicacao)

divisao = array1 / array2
print("Divisão: ",divisao)
```

```python
dados = np.array([10,20,30,40,50])
# Estatísticas Descritivas
media = np.mean(dados)
mediana = np.median(dados)
desvio_padrao = np.std(dados)
minimo = np.min(dados)
maximo = np.max(dados)

print("Média: ",media)
print("Mediana: ",mediana)
print("Desvio Padrão: ",desvio_padrao)
print("Mínimo: ",minimo)
print("Máximo: ",maximo)
```

```python
# Transpor Matrizes
matriz = np.array([[1,2,3],[4,5,6]])

print("Matriz Original")
print(matriz)

matriz_transposta = np.transpose(matriz)

print("Matriz Transposta")
print(matriz_transposta)
```

#Pandas e Numpy

```python
import pandas as pd
import numpy as np

np.random.seed(42) #Garantir a aleatoriedade dos dados

dados_imoveis ={
    'area_m2' : np.random.randint(40, 250, 100),
    'num_quartos' : np.random.randint(1, 5, 100),
    'idade_imovel' : np.random.randint(0, 30, 100),
}

df_imoveis = pd.DataFrame(dados_imoveis)
df_imoveis.head()
```

```python
df_imoveis['preco_mil_reais'] = (
    df_imoveis['area_m2'] * 3.5 + \
    df_imoveis['num_quartos'] * 50 - \
    df_imoveis['idade_imovel'] * 2.5 +
    np.random.normal(0, 40, 100)
).astype(int)

df_imoveis.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_imoveis, x='area_m2', y='preco_mil_reais', hue='num_quartos', palette='viridis', alpha=0.8)
plt.title('Relação entre Área e Preço')
plt.xlabel('Área (m²)')
plt.ylabel('Preço (mil Reais)')
plt.grid(True)
plt.show()
```

```python
caracteristicas = df_imoveis[['area_m2', 'num_quartos', 'idade_imovel']]
alvo = df_imoveis['preco_mil_reais']
```

```python
from sklearn.model_selection import train_test_split # separa uma parte para teste e outra para treino
from sklearn.linear_model import LinearRegression # modelo marrnglearn
from sklearn.metrics import mean_squared_error, r2_score #
```

```python
caracteristicas_treino, caracteristicas_teste, alvo_treino, alvo_teste = train_test_split(caracteristicas, alvo, test_size=0.2, random_state=42)
```

```python
modelo_previsao_preco = LinearRegression()
modelo_previsao_preco.fit(caracteristicas_treino, alvo_treino)
```

```python
print("Coeficientes: ", modelo_previsao_preco.coef_)
print("Intercepto: ", modelo_previsao_preco.intercept_)
```

```python
previsao = modelo_previsao_preco.predict(caracteristicas_teste)
acuracia = modelo_previsao_preco.score(caracteristicas_teste, alvo_teste)
print("Acurácia: ", acuracia)
```

```python
novo_imovel = pd.DataFrame(
    {'area_m2': [150],
     'num_quartos': [2],
     'idade_imovel': [30]
     })
preco_previsto = modelo_previsao_preco.predict(novo_imovel)
print("Preço previsto: ", preco_previsto[0])
```

# Explorando Dados

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
print("--- Carregando e Explorando os Dados ---")
cancer_dados = load_breast_cancer()
df_cancer = pd.DataFrame(np.c_[cancer_dados['data'], cancer_dados['target']], columns = np.append(cancer_dados['feature_names'], ['target']))
df_cancer.head()
```

```python
caracteristicas = df_cancer.drop('target', axis = 1)
resultado = df_cancer['target']
```

```python
print("--- Treinando o Modelo ---")
from sklearn.model_selection import train_test_split
caracteristicas_treino, caracteristicas_teste, resultado_treino, resultado_teste = train_test_split(caracteristicas, resultado, test_size=0.2)

print(f"Tamanho do conjunto de treino: {len(caracteristicas_treino)}")
print(f"Tamanho do conjunto de teste: {len(caracteristicas_teste)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
modelo = RandomForestClassifier(n_estimators = 100, max_depth = 10)
modelo
```

```python
modelo.fit(caracteristicas_treino, resultado_treino)
modelo
```

```python
print("--- Avaliar o Modelo ---")
predicao = modelo.predict(caracteristicas_teste)
predicao
```

```python
from sklearn.metrics import accuracy_score
acuracia = accuracy_score(resultado_teste, predicao)
print(f"Acurácia do Modelo: {acuracia * 100 : .2f}%")
```

```python
print("--- Fazer Previsões com novos dados ---")
novo_exame = np.array(caracteristicas[resultado == 0].mean().values)
novo_exame = novo_exame.reshape(1, -1)
novo_exame
```

```python
previsao = modelo.predict(novo_exame)
previsao
```

```python
classe_prevista = cancer_dados.target_names[int(previsao[0])]
print(f"Dados do Novo Exame (simulado com a media dos tumores malignos): ")
print(novo_exame)
print(f"\nPrevisão do Modelo: O tumor é classificado como {classe_prevista} ({previsao[0]})")
probabilidade_previsao = modelo.predict_proba(novo_exame)
print(f"Probabilidade (Maligno ou Benigno): {probabilidade_previsao[0]}")
```

# K-Means

Agrupamento

K-Means é um algoritmo de aprendizado de máquina são supervisionado usado para agrupar os dados por similaridade.

Supervisionado X Não Supervisionado

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x, y = make_blobs(n_samples = 1000, centers = 3, random_state = 42)
```

```python
modelo = KMeans(n_clusters = 3)
modelo.fit(x)
```

```python
modelo.predict(x)
```
